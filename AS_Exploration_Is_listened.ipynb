```python
import pandas as pd

import pandas as pd

path_to_train = './data/train.csv'
path_to_test = './data/test.csv'

df = pd.read_csv(path_to_train, encoding='utf-8', usecols=['user_id', 'ts_listen', 'is_listened', 'listen_type'])

df.tail()
```

```python
by_user = df.sort_values(by=['user_id', 'ts_listen'], axis=0).groupby('user_id')
by_user.get_group(0)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
%pylab inline
pylab.rcParams['figure.figsize'] = (14, 6)
matplotlib.style.use('ggplot')
```

```python
by_user = df.groupby('user_id').agg({'ts_listen':'count'})
len(by_user[by_user['ts_listen'] < 2])
```

```python
def plot_user_listened(user_id, df):
    fig, ax = plt.subplots()
    colors = {0:'green', 1:'blue'}
    grouped = by_user.get_group(user_id).groupby('listen_type')
    for key, group in grouped:
        group.plot(ax=ax, kind='scatter', x='ts_listen', y='is_listened', label=key, color=colors[key])
    plt.show()
    
plot_user_listened(1000, df)
```
