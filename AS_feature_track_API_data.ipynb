```python
import pandas as pd

path_to_train = './data/train.csv'
path_to_test = './data/test.csv'

train = pd.read_csv(path_to_train, encoding='utf-8', usecols=['user_id', 'media_id', 'is_listened'])
test = pd.read_csv(path_to_test, encoding='utf-8', usecols=['user_id', 'media_id'])
test['is_listened'] = 0

df = pd.concat([train, test], ignore_index=True)
df.tail()
```

```python
path_to_track_data = './data/features/all_media_info.csv'
track_info = pd.read_csv(path_to_track_data, encoding='utf-8', usecols=['Id', 'TrackPosition', 'DiskNumber', 'Rank', 'Rank', 'ExplicitLyrics', 'Bpm', 'Gain', 'Isrc', 'Isrc_Country', 'Isrc_Label'])
track_info.head()
```

```python
df_joined = df.merge(track_info, how='left', left_on='media_id', right_on='Id')
del df_joined['Id']
df_joined.head()
```

```python
len(df_joined[df_joined['Bpm'].isnull()])
```

```python
df_joined[df_joined['Gain'].notnull()][df_joined['Gain'].convert_objects(convert_numeric=True).isnull()]
```

```python
df_joined['TrackPosition'] = df_joined['TrackPosition'].fillna(df_joined['TrackPosition'].mean())
df_joined['DiskNumber'] = df_joined['DiskNumber'].fillna(1)
df_joined['Rank'] = df_joined['Rank'].fillna(df_joined['Rank'].mean())
df_joined['ExplicitLyrics'] = df_joined['ExplicitLyrics'].fillna(False)
df_joined['Bpm'] = df_joined['Bpm'].fillna(df_joined['Bpm'].mean())
df_joined['Gain'] = df_joined['Gain'].fillna(df_joined['Gain'].mean())
df_joined['Isrc_Country'] = df_joined['Isrc_Country'].fillna("unk").map(str)
df_joined['Isrc_Label'] = df_joined['Isrc_Label'].fillna("unk").map(str)
df_joined[df_joined['Isrc_Country'].isnull()].head()
```

```python
from sklearn import preprocessing
le_country = preprocessing.LabelEncoder()
le_country.fit(df_joined['Isrc_Country'])
df_joined['Isrc_Country'] = le_country.transform(df_joined['Isrc_Country'])

le_lyrics = preprocessing.LabelEncoder()
le_lyrics.fit(df_joined['ExplicitLyrics'])
df_joined['ExplicitLyrics'] = le_lyrics.transform(df_joined['ExplicitLyrics'])

le_label = preprocessing.LabelEncoder()
le_label.fit(df_joined['Isrc_Label'])
df_joined['Isrc_Label'] = le_label.transform(df_joined['Isrc_Label'])
df_joined.head()
```

```python
del df_joined['Isrc']
```

```python
by_user_mean = df_joined[df_joined['is_listened']==1].groupby('user_id', as_index=False).agg({'Bpm':'mean', 'Gain':'mean'})
by_user_mean.head()
```

```python
len(by_user_mean), len(df_joined['user_id'].unique())
```

```python
df_joined = df_joined.join(by_user_mean, how='left', on='user_id', rsuffix='ByUserMean')
del df_joined['user_idByUserMean']
df_joined['GainByUserMean'].fillna(df_joined['Gain'].mean())
df_joined['BpmByUserMean'].fillna(df_joined['Bpm'].mean())
df_joined.head()
```

```python
df_joined['GainCenter'] = df_joined['Gain'] - df_joined['GainByUserMean']
df_joined['BpmCenter'] = df_joined['Bpm'] - df_joined['BpmByUserMean']
df_joined.head()
```

```python
by_user_country = df_joined[df_joined['is_listened']==1].groupby(['user_id', 'Isrc_Country'], as_index=False).agg({'media_id':'count'})
by_user_country['Isrc_Country_count'] = by_user_country['media_id']
del by_user_country['media_id']
by_user_country.head()
```

```python
df_joined2 = df_joined.merge(by_user_country, how='left', on=['user_id', 'Isrc_Country'])
df_joined2.head()
```

```python
by_user = df_joined[df_joined['is_listened']==1].groupby('user_id', as_index=False).agg({'media_id':'count'})
by_user['Count'] = by_user['media_id']
del by_user['media_id']
by_user.head()
df_joined3 = df_joined2.merge(by_user, how='left', on='user_id')
df_joined3.head()
```

```python
df_joined3['Isrc_Country_count'] = df_joined3['Isrc_Country_count']/df_joined3['Count']
```

```python
path_to_api_track_data_train = './data/features/api_data_track.train.txt'
path_to_api_track_data_test = './data/features/api_data_track.test.txt'
newcols = ['TrackPosition','DiskNumber','Rank','ExplicitLyrics', 
           'Bpm','Gain', 'Isrc_Country','Isrc_Label','BpmByUserMean',
           'GainByUserMean','GainCenter', 'BpmCenter', 'Isrc_Country_count']
```

```python
train_dt = df_joined3.loc[range(len(train))][newcols]
train_dt.to_csv(path_to_api_track_data_train, encoding='utf-8', index=False)

test_dt = df_joined3.loc[range(len(train), len(train)+len(test))][newcols]
test_dt.to_csv(path_to_api_track_data_test, encoding='utf-8', index=False)
```
