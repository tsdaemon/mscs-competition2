```python
from graph_tool.all import *
import pandas as pd

path_to_train = './data/train.csv'

train = pd.read_csv(path_to_train, encoding='utf-8')
train.head()
```

```python
g = Graph(directed=False)
artists = {}
art_to_album = {}
albums = {}
alb_to_media = {}
medias = {}
users = {}

vprop_id = g.new_vertex_property("int")
g.vertex_properties["id"] = vprop_id

vprop_label = g.new_vertex_property("string")
g.vertex_properties["label"] = vprop_label

eprop_label = g.new_edge_property("string")
g.edge_properties["label"] = eprop_label

eprop_ts = g.new_edge_property("int")
g.edge_properties["ts"] = eprop_ts

processed = 0

def convert_row_to_graph(row):
    global g, artists, albums, medias, users, vprop_id, vprop_label, eprop_label, art_to_album, alb_to_media, processed
    
    if(row['user_id'] not in users):
        u = g.add_vertex()
        users[row['user_id']] = u
        vprop_id[u] = row['user_id']
        vprop_label[u] = 'USER'
        
    user = users[row['user_id']]
    
    if(row['artist_id'] not in artists):
        art = g.add_vertex()
        artists[row['artist_id']] = art
        vprop_id[art] = row['artist_id']
        vprop_label[art] = 'ARTIST'
    
    artist = artists[row['artist_id']]
    
    if(row['album_id'] not in albums):
        alb = g.add_vertex()
        albums[row['album_id']] = alb
        vprop_id[alb] = row['album_id']
        vprop_label[alb] = 'ALBUM'
        
    album = albums[row['album_id']]
    
    if((row['artist_id'], row['album_id']) not in art_to_album):
        a_to_a = g.add_edge(artist, album)
        art_to_album[(row['artist_id'], row['album_id'])] = a_to_a
        eprop_label[a_to_a] = 'CREATION'
    
    if(row['media_id'] not in medias):
        m = g.add_vertex()
        medias[row['media_id']] = m
        vprop_id[m] = row['media_id']
        vprop_label[m] = 'MEDIA'
    
    media = medias[row['media_id']]
    
    if((row['album_id'], row['media_id']) not in alb_to_media):
        a_to_m = g.add_edge(album, media)
        alb_to_media[(row['album_id'], row['media_id'])] = a_to_m
        eprop_label[a_to_m] = 'CREATION'
        
    if(row['is_listened']==1):
        is_listened = g.add_edge(user, media)
        eprop_label[is_listened] = 'LISTENED'
        eprop_ts[is_listened] = row['ts_listen']
    
    processed += 1
    if(processed%10000 == 0):
        print("Done: %d/%d"%(processed, len(train)))
        
train.apply(convert_row_to_graph, axis=1)
```

```python
g.save("./data/graph.xml.gz")
```

```python
from graph_tool.all import *
g = load_graph("./data/graph.xml.gz")
```

```python
import pandas as pd
path_to_train = './data/train.csv'

train = pd.read_csv(path_to_train, encoding='utf-8')
train.head()
```

```python
# first create indexes
artists = {}
albums = {}
medias = {}
users = {}

vprop_id = g.vertex_properties["id"]
vprop_label = g.vertex_properties["label"]

for v in g.vertices():
    label = vprop_label[v]
    id = vprop_id[v]
    if label == "USER":
        users[id] = v
    elif label == "MEDIA":
        medias[id] = v
    elif label == "ALBUM":
        albums[id] = v
    elif label == "ARTIST":
        artists[id] = v
```

```python
# first lets try general shortest distance
eprop_label = g.edge_properties["label"]
eprop_ts = g.edge_properties["ts"]

def shortest(g, row):
    media = medias[row['media_id']]
    user = users[row['user_id']]
    ts = row['ts_listen']
    g.set_edge_filter(eprop_ts < ts)
    print("Start search")
    dist = shortest_distance(g, user, media, max_dist=20)
    print(dist)
    return len(dist)
    

distance_to_media = train.apply(lambda row: shortest(g, row),axis=1)
```
