```python
import pandas as pd

path_to_train = './data/train.csv'
path_to_test = './data/test.csv'

train = pd.read_csv(path_to_train, encoding='utf-8', usecols=['user_id', 'ts_listen',  'is_listened'])
test = pd.read_csv(path_to_test, encoding='utf-8', usecols=['user_id', 'ts_listen'])
test['is_listened'] = 0 # просто щоб зібрати, ми будемо використовувати попередні значення

df = pd.concat([train, test], ignore_index=True)
df.tail()
```

```python
df = df.sort_values(by=['user_id', 'ts_listen'], axis=0)
df.tail()
```

```python
by_user_count = df.groupby('user_id', as_index=False).agg({'ts_listen':'count'})
_2records_users = by_user_count[by_user_count['ts_listen'] == 2]['user_id']
```

```python
df['is_listened_prev'] = df['is_listened'].shift(1)
fi = df.groupby('user_id', as_index=False).apply(lambda x: x.index[0])
df.ix[fi, 'is_listened_prev'] = -100

df['is_listened_prev2'] = df['is_listened'].shift(2)
fi2 = df.groupby('user_id', as_index=False).apply(lambda x: x.index[1])
df.ix[fi2, 'is_listened_prev2'] = -100
df.ix[fi, 'is_listened_prev2'] = -100

df['is_listened_prev3'] = df['is_listened'].shift(3)
fi3 = df[~df['user_id'].isin(_2records_users)].groupby('user_id', as_index=False).apply(lambda x: x.index[2])
df.ix[fi3, 'is_listened_prev3'] = -100
df.ix[fi2, 'is_listened_prev3'] = -100
df.ix[fi, 'is_listened_prev3'] = -100
df.ix[df['user_id'].isin(_2records_users), 'is_listened_prev3'] = -100

df.head()
```

```python
df.tail()
```

```python
df = df.sort_index()
df.tail()
```

```python
path_to_test_prev_is_listened = './data/features/prev_is_listened.test.csv'
path_to_train_prev_is_listened = './data/features/prev_is_listened.train.csv'
new_cols = [
    'is_listened_prev', 
    'is_listened_prev2', 
    'is_listened_prev3'
]
```

```python
test_cumsum = df.iloc[range(len(train), len(train)+len(test))][new_cols]
test_cumsum.to_csv(path_to_test_prev_is_listened, encoding='utf-8', index=False)

train_cumsum = df.iloc[range(len(train))][new_cols]
train_cumsum.to_csv(path_to_train_prev_is_listened, encoding='utf-8', index=False)
```

```python
import pandas as pd

path_to_train = './data/train.csv'
path_to_test = './data/test.csv'

train = pd.read_csv(path_to_train, encoding='utf-8', usecols=['user_id', 'ts_listen',  'is_listened'])
test = pd.read_csv(path_to_test, encoding='utf-8', usecols=['user_id', 'ts_listen'])
test['is_listened'] = 0 # просто щоб зібрати, ми будемо використовувати попередні значення

df = pd.concat([train, test], ignore_index=True)
df.tail()
```

```python
df = df.sort_values(by=['user_id', 'ts_listen'], axis=0)
df.tail()
```

```python
df['is_listened_prev'] = df['is_listened'].shift(1)
df['ts_listen_prev'] = df['ts_listen'].shift(1)

fi = df.groupby('user_id', as_index=False).apply(lambda x: x.index[0])
df.ix[fi, 'is_listened_prev'] = 0
df.ix[fi, 'ts_listen_prev'] = 0

df.head()
```

```python
time_window = 5*60 #5 minutes
df['was_is_listened_before'] = ((df['ts_listen']-df['ts_listen_prev']) < time_window * df['is_listened_prev']).astype(int)
df.head(20)
```

```python
df = df.sort_index()
df.tail()
```

```python
from scipy.stats.stats import pearsonr 

pearsonr(df['is_listened'], df['was_is_listened_before'])
```

```python
path_to_test_prev_is_listened = './data/features/prev_is_listened.test.csv'
path_to_train_prev_is_listened = './data/features/prev_is_listened.train.csv'
new_cols = [
    'was_is_listened_before'
]
```

```python
test_cumsum = df.iloc[range(len(train), len(train)+len(test))][new_cols]
test_cumsum.to_csv(path_to_test_prev_is_listened, encoding='utf-8', index=False)

train_cumsum = df.iloc[range(len(train))][new_cols]
train_cumsum.to_csv(path_to_train_prev_is_listened, encoding='utf-8', index=False)
```

```python
import validation as v
import pandas as pd

path_to_train = './data/train.csv'

train = pd.read_csv(path_to_train, encoding='utf-8')
train_prev_is_listened = pd.read_csv(path_to_train_prev_is_listened, encoding='utf-8')

train = pd.concat([train, train_prev_is_listened], axis=1)
train.head()
```

```python
path_to_folds = './data/lightgbm/prev_is_listened'
```

```python
v.generate_validation_folds(train, path_to_folds)
```

```python
import model_lgbm as mlgbm

cols = [
    'genre_id', 
    #'media_id', 
    #'album_id', 
    'context_type',
    'release_date', 
    'platform_name', 
    'platform_family', 
    'media_duration',
    'listen_type',
    'user_gender', 
    'user_id',
    #'artist_id', 
    'user_age'
] + new_cols

params = {
    'application':'binary',
    'num_leaves': 24,
    'max_depth': 15,
    'feature_fraction': 0.7,
    'bagging_fraction': 0.7,
    'max_bin': 80,
    'metric': 'auc',
    'verbose': 1
}
```

```python
evs = mlgbm.crossvalidate_model(path_to_folds,cols,params,validate_on_both=True, iterations=100)
```

```python
mlgbm.feature_score(evs[1],cols)
```

```python
path_to_sample = './data/sample_submission_kaggle.csv'
path_to_submission = './data/submission.csv'
path_to_test = './data/test.csv'

test = pd.read_csv(path_to_test, encoding='utf-8')
test = pd.concat([test, pd.read_csv(path_to_test_prev_is_listened, encoding='utf-8')], axis=1)

test.head()
```

```python
fs = mlgbm.create_submission(train, test, path_to_sample, params, cols, path_to_submission, iterations=400)
```

```python
mlgbm.feature_score(fs,cols)
```
